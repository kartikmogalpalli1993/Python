```python
# ============================================================
# RENT vs HOME LOAN CALCULATOR
# Premium Financial Dashboard UI
# IDE: Jupyter Notebook
# ============================================================

from IPython.display import display, HTML
import ipywidgets as widgets
import math

# ============================================================
# GLOBAL UI DESIGN
# ============================================================

display(HTML("""

<style>

body{
    background:#0B1120;
}

/* Main Title */

.main-title{
    font-size:52px;
    font-weight:800;
    font-family:Segoe UI;
    background:linear-gradient(90deg,#7C3AED,#2563EB);
    -webkit-background-clip:text;
    -webkit-text-fill-color:transparent;
    margin-bottom:10px;
}

.sub-title{
    color:#94A3B8;
    font-size:18px;
    font-family:Segoe UI;
    margin-bottom:30px;
}

/* Cards */

.card{
    background:linear-gradient(145deg,#111827,#1E293B);
    border-radius:28px;
    padding:30px;
    margin-top:25px;
    box-shadow:0px 10px 40px rgba(0,0,0,0.45);
    border:1px solid rgba(255,255,255,0.08);
}

/* Metric Cards */

.metric-purple{
    background:linear-gradient(135deg,#4F46E5,#7C3AED);
    border-radius:24px;
    padding:25px;
    color:white;
    margin-bottom:20px;
}

.metric-green{
    background:linear-gradient(135deg,#059669,#10B981);
    border-radius:24px;
    padding:25px;
    color:white;
    margin-bottom:20px;
}

.metric-blue{
    background:linear-gradient(135deg,#2563EB,#3B82F6);
    border-radius:24px;
    padding:25px;
    color:white;
    margin-bottom:20px;
}

.metric-title{
    font-size:17px;
    opacity:0.9;
    font-family:Segoe UI;
}

.metric-value{
    font-size:42px;
    font-weight:800;
    margin-top:12px;
    font-family:Segoe UI;
}

/* Summary */

.summary-row{
    display:flex;
    justify-content:space-between;
    padding:15px 0px;
    border-bottom:1px solid rgba(255,255,255,0.08);
    color:white;
    font-size:17px;
    font-family:Segoe UI;
}

.green{
    color:#10B981;
    font-weight:700;
}

.red{
    color:#EF4444;
    font-weight:700;
}

.blue{
    color:#60A5FA;
    font-weight:700;
}

.purple{
    color:#A78BFA;
    font-weight:700;
}

.section-title{
    color:white;
    font-size:28px;
    font-weight:700;
    margin-bottom:25px;
    font-family:Segoe UI;
}

.footer-note{
    color:#94A3B8;
    margin-top:20px;
    font-size:14px;
    font-family:Segoe UI;
}

</style>

"""))

# ============================================================
# HEADER
# ============================================================

display(HTML("""

<div class='main-title'>
🏠 Rent vs Home Loan Calculator
</div>

<div class='sub-title'>
Compare Renting vs Buying a House with Interactive Financial Analysis
</div>

"""))

# ============================================================
# INPUTS
# ============================================================

style = {'description_width':'220px'}
layout = widgets.Layout(width='700px')

# Property Inputs
property_value = widgets.FloatSlider(
    value=8000000,
    min=1000000,
    max=50000000,
    step=100000,
    description='Property Value',
    style=style,
    layout=layout
)

down_payment = widgets.FloatSlider(
    value=2000000,
    min=0,
    max=20000000,
    step=50000,
    description='Down Payment',
    style=style,
    layout=layout
)

interest_rate = widgets.FloatSlider(
    value=8,
    min=1,
    max=15,
    step=0.1,
    description='Interest Rate (%)',
    style=style,
    layout=layout
)

loan_years = widgets.IntSlider(
    value=20,
    min=1,
    max=30,
    step=1,
    description='Loan Tenure (Years)',
    style=style,
    layout=layout
)

monthly_rent = widgets.FloatSlider(
    value=22000,
    min=5000,
    max=200000,
    step=1000,
    description='Monthly Rent',
    style=style,
    layout=layout
)

rent_increase = widgets.FloatSlider(
    value=5,
    min=1,
    max=20,
    step=0.5,
    description='Annual Rent Increase (%)',
    style=style,
    layout=layout
)

# ============================================================
# BUTTON
# ============================================================

calculate_btn = widgets.Button(
    description='Analyze Financial Decision',
    button_style='success',
    icon='home',
    layout=widgets.Layout(width='350px', height='55px')
)

output = widgets.Output()

# ============================================================
# CALCULATOR LOGIC
# ============================================================

def calculate_home_loan(principal, annual_rate, years):

    monthly_rate = annual_rate / 12 / 100
    months = years * 12

    emi = (
        principal * monthly_rate *
        ((1 + monthly_rate) ** months)
    ) / (((1 + monthly_rate) ** months) - 1)

    return emi


def generate_dashboard(b):

    output.clear_output()

    # ========================================================
    # HOME LOAN
    # ========================================================

    loan_amount = property_value.value - down_payment.value

    emi = calculate_home_loan(
        loan_amount,
        interest_rate.value,
        loan_years.value
    )

    total_home_cost = emi * loan_years.value * 12

    # ========================================================
    # RENT COST
    # ========================================================

    total_rent = 0
    current_rent = monthly_rent.value

    for year in range(loan_years.value):

        yearly_rent = current_rent * 12
        total_rent += yearly_rent

        current_rent = current_rent * (
            1 + rent_increase.value/100
        )

    # ========================================================
    # SAVINGS
    # ========================================================

    savings = total_home_cost - total_rent

    if savings > 0:
        recommendation = "Renting is Better"
        color_class = "green"
    else:
        recommendation = "Buying is Better"
        color_class = "red"

    # ========================================================
    # HTML DASHBOARD
    # ========================================================

    html = f"""

    <div class='card'>

        <div class='section-title'>
        📊 Financial Decision Summary
        </div>

        <div class='metric-purple'>

            <div class='metric-title'>
            {recommendation}
            </div>

            <div class='metric-value'>
            ₹ {abs(savings):,.0f}
            </div>

            <div style='margin-top:10px;font-size:16px;'>
            Difference over {loan_years.value} years
            </div>

        </div>

    </div>


    <div class='card'>

        <div class='section-title'>
        🏠 Home Loan Analysis
        </div>

        <div class='metric-green'>

            <div class='metric-title'>
            Monthly EMI
            </div>

            <div class='metric-value'>
            ₹ {emi:,.0f}
            </div>

        </div>

        <div class='summary-row'>
            <span>Property Value</span>
            <span class='blue'>
            ₹ {property_value.value:,.0f}
            </span>
        </div>

        <div class='summary-row'>
            <span>Down Payment</span>
            <span class='purple'>
            ₹ {down_payment.value:,.0f}
            </span>
        </div>

        <div class='summary-row'>
            <span>Loan Amount</span>
            <span class='green'>
            ₹ {loan_amount:,.0f}
            </span>
        </div>

        <div class='summary-row'>
            <span>Total EMI Paid</span>
            <span class='red'>
            ₹ {total_home_cost:,.0f}
            </span>
        </div>

    </div>



    <div class='card'>

        <div class='section-title'>
        🏡 Rental Analysis
        </div>

        <div class='metric-blue'>

            <div class='metric-title'>
            Monthly Rent
            </div>

            <div class='metric-value'>
            ₹ {monthly_rent.value:,.0f}
            </div>

        </div>

        <div class='summary-row'>
            <span>Initial Monthly Rent</span>
            <span class='blue'>
            ₹ {monthly_rent.value:,.0f}
            </span>
        </div>

        <div class='summary-row'>
            <span>Annual Rent Increase</span>
            <span class='purple'>
            {rent_increase.value}%
            </span>
        </div>

        <div class='summary-row'>
            <span>Total Rent Paid</span>
            <span class='green'>
            ₹ {total_rent:,.0f}
            </span>
        </div>

        <div class='summary-row'>
            <span>Comparison Period</span>
            <span class='red'>
            {loan_years.value} Years
            </span>
        </div>

    </div>



    <div class='card'>

        <div class='section-title'>
        📌 Recommendation
        </div>

        <div class='summary-row'>
            <span>Financially Better Option</span>
            <span class='{color_class}'>
            {recommendation}
            </span>
        </div>

        <div class='summary-row'>
            <span>Estimated Savings Difference</span>
            <span class='green'>
            ₹ {abs(savings):,.0f}
            </span>
        </div>

        <div class='footer-note'>
        ✔ This analysis compares cumulative rent payments against total home loan EMI outflow.
        </div>

    </div>

    """

    with output:
        display(HTML(html))

# ============================================================
# BUTTON EVENT
# ============================================================

calculate_btn.on_click(generate_dashboard)

# ============================================================
# DISPLAY INPUTS
# ============================================================

display(HTML("<div class='section-title'>⚙ Property Details</div>"))

display(property_value)
display(down_payment)

display(HTML("<div class='section-title'>🏦 Loan Parameters</div>"))

display(interest_rate)
display(loan_years)

display(HTML("<div class='section-title'>🏡 Rental Details</div>"))

display(monthly_rent)
display(rent_increase)

display(HTML("<br>"))

display(calculate_btn)

display(output)
```
