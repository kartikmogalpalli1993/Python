```python
# =========================================================
# POLARS vs PANDAS PERFORMANCE BENCHMARK
# GOOGLE COLAB VERSION
# =========================================================

# ---------------------------------------------------------
# STEP 1 : INSTALL LIBRARIES
# ---------------------------------------------------------

!pip install pandas polars pyarrow -q


# ---------------------------------------------------------
# STEP 2 : IMPORT LIBRARIES
# ---------------------------------------------------------

import pandas as pd
import polars as pl
import time


# ---------------------------------------------------------
# STEP 3 : FILE PATH
# GOOGLE COLAB FILE LOCATION
# ---------------------------------------------------------

# AFTER UPLOADING FILE IN COLAB
# FILE WILL BE AVAILABLE INSIDE /content/

csv_file = "/content/spotify_user_behavior_realistic_50000_rows.csv"

print("\n📂 CSV FILE PATH :")
print(csv_file)


# =========================================================
# PANDAS EXECUTION
# =========================================================

try:

    start = time.time()

    pandas_df = pd.read_csv(csv_file)

    # Sample Operation
    pandas_result = pandas_df.head(5000)

    pandas_time = time.time() - start

    print("\n" + "=" * 60)
    print("📘 PANDAS DATAFRAME")
    print("=" * 60)

    print(pandas_df.head())

    print("\n📘 Shape :", pandas_df.shape)

    print("📘 Execution Time :",
          round(pandas_time, 4), "seconds")

except Exception as e:

    print("\n❌ Pandas Error :", e)

    pandas_time = None


# =========================================================
# POLARS EXECUTION
# =========================================================

try:

    start = time.time()

    polars_df = pl.read_csv(csv_file)

    # Sample Operation
    polars_result = polars_df.head(5000)

    polars_time = time.time() - start

    print("\n" + "=" * 60)
    print("⚡ POLARS DATAFRAME")
    print("=" * 60)

    print(polars_df.head())

    print("\n⚡ Shape :", polars_df.shape)

    print("⚡ Execution Time :",
          round(polars_time, 4), "seconds")

except Exception as e:

    print("\n❌ Polars Error :", e)

    polars_time = None


# =========================================================
# PERFORMANCE RESULT
# =========================================================

print("\n" + "=" * 60)
print("🚀 PERFORMANCE RESULT")
print("=" * 60)

if pandas_time and polars_time:

    if pandas_time < polars_time:

        winner = "Pandas"
        speed = polars_time / pandas_time

    else:

        winner = "Polars"
        speed = pandas_time / polars_time

    print(f"\n🏆 Faster Library : {winner}")

    print(f"🚀 Speed Difference : {round(speed, 2)}x Faster")

    print("\n📊 EXECUTION SUMMARY")
    print("-" * 40)

    print("📘 Pandas :", round(pandas_time, 4), "seconds")

    print("⚡ Polars :", round(polars_time, 4), "seconds")

else:

    print("\n❌ Unable to Compare")


# =========================================================
# IMPORTANT FOR GOOGLE COLAB
# =========================================================

# 1. LEFT PANEL -> FILES
# 2. CLICK UPLOAD
# 3. UPLOAD:
#    spotify_user_behavior_realistic_50000_rows.csv
#
# FILE PATH WILL BE:
# /content/spotify_user_behavior_realistic_50000_rows.csv
#
# =========================================================
```
