```python
# Make sure this function exists
import hashlib

def encrypt_password(password):
    return hashlib.sha256(password.encode()).hexdigest()


# Make sure final_passwords exists
import re

name_clean = re.sub(r"\s+", "", full_name).lower()
day, month, year = dob.split("-")

combinations = [
    name_clean,
    name_clean + day,
    name_clean + year,
    name_clean + day + month,
    name_clean.capitalize() + "@" + year,
    name_clean[:3] + day + "!",
    name_clean + "123",
    name_clean.capitalize() + "#" + day + month
]

def apply_rules(password, instructions):
    if "uppercase" in instructions.lower():
        password = password.capitalize()
    if "numbers" in instructions.lower() and not any(c.isdigit() for c in password):
        password += "123"
    if "special" in instructions.lower() and not any(c in "!@#$%" for c in password):
        password += "@"
    if "min" in instructions.lower() and len(password) < 8:
        password += "X" * (8 - len(password))
    return password

final_passwords = [apply_rules(p, instructions) for p in combinations]
```

```python
#Step 6: Encrypt (Hash)
encrypted_passwords = [encrypt_password(p) for p in final_passwords]
```

```python
#Step 7: Create DataFrame
df = pd.DataFrame({
    "Full_Name": [full_name]*len(final_passwords),
    "DOB": [dob]*len(final_passwords),
    "Instructions": [instructions]*len(final_passwords),
    "Generated_Password": final_passwords,
    "Encrypted_Password": encrypted_passwords
})
```

```python
df.to_csv("passwords.csv", mode="w", index=True, header=True)
```
