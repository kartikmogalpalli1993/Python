```python
##select the Required Columns on Dataframe 
from IPython.display import display
import pandas as pd

df = pd.read_csv("odb.csv")
selected_df["Player"] = df["Player"].str.replace(r"[^a-zA-Z\s]", "", regex=True)
##selected_df = df[[ "Span","Mat","Inns","NO","Runs","HS","Ave","BF","SR","100","50","0","4s","6s"] ]  

selected_df = selected_df.copy()

# Step 1: normalize spaces (VERY IMPORTANT)
selected_df["Player"] = selected_df["Player"].str.replace(r"\s+", " ", regex=True).str.strip()

# Step 2: split correctly
selected_df[["Player_Name", "Country"]] = selected_df["Player"].str.rsplit(" ", n=1, expand=True)

display(selected_df.head())
```

```python
selected_df = selected_df.copy()

# Split last word as country
selected_df[["Player_Name", "Country"]] = selected_df["Player"].str.rsplit(" ", n=1, expand=True)

display(selected_df.head())
```

```python
selected_df = selected_df.drop(columns=["Player"])
```

```python
display(selected_df.head())
```

```python
#To make the columns in Order Wise
cols = ["Player_Name", "Country"] + [col for col in selected_df.columns if col not in ["Player_Name", "Country"]]

selected_df = selected_df[cols]

display(selected_df.head())
```

```python
selected_df["Country"] = selected_df["Country"].replace({
    "AUSICC": "AUS",
    "AsiaSL": "SL",
    "AfrSA":"SA",
    "AsiaICCSL)":"SL",
    "AfrICCSA":"SA",
    "AsiaPAK":"PAK",
    "AsiaICCINDIA":"INDIA"
})

display(selected_df.head())
```

```python
##To change Country AUS to Australia
selected_df["Country"] = selected_df["Country"].replace({
    "AUS": "Australia",
    "INDIA": "India",
    "SL": "Sri Lanka",
    "SA": "South Africa",
    "PAK": "Pakistan"
})

display(selected_df.head())
```
